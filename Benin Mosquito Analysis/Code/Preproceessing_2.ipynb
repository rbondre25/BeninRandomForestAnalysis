```python
def clean_data(fileName):
  df = pd.read_csv(io.BytesIO(uploaded[fileName]))
  df.drop([0], axis = 0, inplace = True)
  df.date = df.date.apply(lambda x: dt.datetime.strptime(x, "%Y-%m-%d").date())
  mask = df.date < dt.date(2017, 5, 30)
  df = df.loc[~mask]

  return df
```

```python
df['magnitude'] = 0
#0-10 ==> 0
#11-50 ==> 1
#51 << ==> 2

def label(n):
  l = 0
  if int(n) < 9:
    l = 0
  if int(n) > 9:
    l = 1
  if int(n) > 22:
    l = 2
  return l


```
